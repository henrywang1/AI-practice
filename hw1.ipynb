## (b.i)
The implementation of K-NN is shown as below.


```
from collections import Counter
import random
from scipy import stats

class KNNClassifier(object):
    def fit(self, x, y):
        self.x_train = x
        self.y_train = y

    def pred(self, x, k, dist_func="euclidean"):
        def euclidean_dist(a, b):
            dist = np.sum(np.square(a)[:,np.newaxis,:], axis=2) \
             - 2 * a.dot(b.T) \
             + np.sum(np.square(b), axis=1)
            dist = np.sqrt(dist)
            return dist

        def l1norm_dist(a, b):
            dist = a[:,np.newaxis,:] - b
            dist = abs(dist).sum(2)
            return dist
        
        def cosine_dist(a, b):            
            a = a / np.sqrt((a * a).sum(axis=1))[:,np.newaxis]
            b = b / np.sqrt((b * b).sum(axis=1))[:,np.newaxis]
            return 1 - a.dot(b.T)
    
        if dist_func == "euclidean":
            dist = euclidean_dist(x, self.x_train)
        elif dist_func == "l1norm":
            dist = l1norm_dist(x, self.x_train)
        elif dist_func == "cosine":
            dist = cosine_dist(x, self.x_train)
        else:
            raise ValueError("dist_func must be euclidean, l1norm or cosine, {} is not support".format(dist_func))

        idx = dist.argpartition(k)
        idx = idx[:,:k]        
        dist = dist[np.arange(dist.shape[0])[:, None], idx]
        y_pred = self.y_train[idx]
        y_pred = (stats.mode(y_pred, axis=1)[0].flatten())
        
        return y_pred, dist, idx
```

## (b.ii)
$K^* = 1$

```
x_train = df_train[features].to_numpy()
y_train = df_train["class"].to_numpy()

x_val = df_val[features].to_numpy()
y_val = df_val["class"].to_numpy()

knn = KNNClassifier()
knn.fit(x_train, y_train)

results = []
k_star = 1
acc_star = 0
for k in range(1, 21, 2):
    y_pred, dist, idx = knn.pred(x_val, k)
    acc = (y_pred == y_val).mean().round(4)
    if acc > acc_star:
        acc_star = acc
        k_star = k
    results.append({"Acc": acc*100, "K": k})

df = pd.DataFrame(results)

rcParams['figure.figsize']=(8,5)
rcParams['font.size'] = 14
ax = sns.lineplot(x="K", y="Acc", data=df, marker='o')
ax.set_xticks(df.K)
ax.set(ylabel="Accuracy (%)")
plt.plot()
print("K*={}".format(k_star))
```

## (b.iv)
* I tried l1-norm, cosine, and euclidean distance on the development and test set under different $k$
* The result shows that cosine distance has the best performance on both development and test set
* With $K=1$, the KNN algorithm has the best performance under all the distance metrics

```
metrics = ["euclidean", "l1norm", "cosine"]
rcParams['figure.figsize']=(8,12)
rcParams['font.size'] = 14
fig = plt.figure()
fig.subplots_adjust(hspace=0.4)

for i in range(1, 3):
    plt.subplot(2, 1, i)
    x = x_val if i == 1 else x_test
    y = y_val if i == 1 else y_test
    results = []
    for metric in metrics:
        for k in range(1, 21, 2):
            y_pred, dist, idx = knn.pred(x, k, metric)
            acc = (y_pred == y).mean().round(4)
            results.append({"Acc": acc*100, "K": k, "Metric": metric})

    df = pd.DataFrame(results)
    ax = sns.lineplot(x="K", y="Acc", data=df, hue="Metric", legend=('full'if i==1 else False))
    ax.set_xticks(df.K)
    ax.set(ylabel="Accuracy (%)")
    if i == 1:
        ax.legend(loc='center left', bbox_to_anchor=(1.1, -0.3), fontsize=14, ncol=1)
        ax.set_title("(a) Result on the development set", y=-0.3, fontweight='bold', fontsize=14)
    else:
        ax.set_title("(b) Result on the test set", y=-0.3, fontweight='bold', fontsize=14)
plt.plot()
```
